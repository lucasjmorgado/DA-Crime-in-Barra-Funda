```python
import pandas as pd

# File paths to your Excel files
excel_files = [
    '/content/SPDadosCriminais_2024.csv',
    '/content/SPDadosCriminais_2023.csv',
    '/content/SPDadosCriminais_2022.csv'
]  # Replace with your file paths

# Initialize a list to hold DataFrames
dataframes = []

# Load the necessary columns from each file
for file_name in excel_files:
    df = pd.read_csv(file_name, usecols=[
        'CIDADE',
        'DATA_OCORRENCIA_BO',
        'HORA_OCORRENCIA_BO',
        'BAIRRO',
        'LOGRADOURO',
        'NUMERO_LOGRADOURO',
        'LATITUDE',
        'LONGITUDE',
        'NATUREZA_APURADA'
    ])
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
concat_df = pd.concat(dataframes, ignore_index=True)

# Display the concatenated DataFrame
concat_df.head()
```

```python
# Create backup df
bkp_df = concat_df.copy()
bkp_df.head()
```

```python
# Check kinds of crime
print(concat_df['NATUREZA_APURADA'].unique())
```

```python
# Create dataframe with only robbery data and desired city and neighborhood
crime_barrafunda_roubos = concat_df.loc[(concat_df['BAIRRO'] == 'BARRA FUNDA')
                                  & (concat_df['CIDADE'] == 'S.PAULO')
                                  & ((concat_df['NATUREZA_APURADA'] == 'ROUBO - OUTROS')  | (concat_df['NATUREZA_APURADA'] == 'LATROCÍNIO'))]
# Clean empty location data
crime_barrafunda_roubos = crime_barrafunda_roubos.dropna(subset=['LATITUDE', 'LONGITUDE'])
crime_barrafunda_roubos.head(500)
```

```python
# Create dataframe with only theft data and desired city and neighborhood
crime_barrafunda_furtos = concat_df.loc[(concat_df['BAIRRO'] == 'BARRA FUNDA')
                                  & (concat_df['CIDADE'] == 'S.PAULO')
                                  & ((concat_df['NATUREZA_APURADA'] == 'FURTO - OUTROS')  | (concat_df['NATUREZA_APURADA'] == 'LATROCÍNIO'))]
# Clean empty location data
crime_barrafunda_furtos = crime_barrafunda_furtos.dropna(subset=['LATITUDE', 'LONGITUDE'])
crime_barrafunda_furtos.head(500)
```

```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster

# Create a map
map_barra_funda = folium.Map(location=[-23.5228831,-46.6670507], tiles='cartodbpositron', zoom_start=16)

# Create markers for each robbery
for idx, row in crime_barrafunda_roubos.iterrows():
    Marker([row['LATITUDE'], row['LONGITUDE']]).add_to(map_barra_funda)

map_barra_funda
```

```python
import folium
from folium.plugins import HeatMap

# Create a map centered around the specified coordinates
map_barra_funda = folium.Map(location=[-23.5228831, -46.6670507], tiles='cartodbpositron', zoom_start=15)

# Extracting latitude and longitude data from the DataFrame
heat_data = [[row['LATITUDE'], row['LONGITUDE']] for index, row in crime_barrafunda_roubos.iterrows()]

# Create a HeatMap and add it to the map
HeatMap(heat_data, radius=10).add_to(map_barra_funda)

# Display the map
map_barra_funda
```
